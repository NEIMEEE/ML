Importing the Dependecies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection + Pre-Processing

```python
# loading the data from csv file to a pandas Dataframe
raw_mail_data = pd.read_csv('/content/mail_data.csv')
```

```python
print(raw_mail_data)
```

```python
# replace the null values with a null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
# printing the first 5 rows of the dataframe
mail_data.head()
```

```python
#checking the number of raws and coloumns int the dataframe
mail_data.shape
```

Label Encoding

spam-0




ham-1



```python
#label spam mail as 0 ; ham mail as 1 ;

mail_data.loc[mail_data['Category'] == 'spam', 'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',] = 1
```

```python
# separatung the data as texts and label

X = mail_data['Message']
Y = mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

Spilitting the Sata into Training Data and Test Data

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

Feature  Extraction ( conerting the dta into numerical value )

```python
# transform the text data to feature vectors that can be used as input to the ogistic regression

feature_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# Convert Y-Train and Y_test valus as integers

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_features)
```

Training the Model : Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logitic Regression model with the training data
model.fit(X_train_features, Y_train)
```

Evaluating the traned model

```python
# prediction n training data
prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Accuracy on training data : ' , accuracy_on_training_data)
```

```python
# prediction n training data

prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print('Accuracy on test data : ' , accuracy_on_test_data)
```

Building a Predective System


```python
input_mail = ["As per your request 'Melle Melle (Oru Minnaminunginte Nurungu Vettam)' has been set as your callertune for all Callers. Press *9 to copy your friends Callertune"]

# convert text to feature vectors
input_data_features = feature_extraction.transform(input_mail)

# making prediction

prediction = model.predict(input_data_features)
print(prediction)

if (prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
```
